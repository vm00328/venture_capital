```python
# Imports & Packages
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt

pd.set_option('display.max_rows', 200)
pd.set_option('display.max_columns', None)

data = pd.read_excel("C:/Users/VladislavManolo/project/data/Historical Performance/historical_MOIC_data_FINAL.xlsx")
print(len(data))

data = data.dropna(subset = data.columns[data.columns.get_loc('CQ12000'):], how = 'all')
print("Number of funds containing at least 1 historical performance data point: ", len(data))

last_quarter_column = data.columns[data.columns.str.contains('CQ')].max() # Find the last column with non-NaN values
print(last_quarter_column, type(last_quarter_column))
```

**For each fund:**

- create X columns where X is the number of years that have passed between the fund's vintage and its last reported date (that is the **Fund Age**) (the data is reported on a quarterly basis so we need to find the last column with quarterly data containing non-NaN values)
  
- in each of those columns take the average return from the quarters of the given year

- then take the data from the latest year and based on it assign quartile rankings to the funds (1st, 2nd, 3rd, 4th)

- For all of the funds within each quartile group calculate the average performance for each year of the funds' lifecycle

- Create 4 line charts (one for each quartile) showing the average return per year (the return is on the y-axis) for each year of the funds' lifecycle (the years denoting the lifecycle of a fund such as 1,2,3,4...10 are on the x-axis)

```python
data['Latest Data Year'] = 2019

data['Fund Age'] = (data['Latest Data Year'] - data['VINTAGE'] + 1)

# Creating X extra columns for every fund where x is equal to the age of the fund
for i in range(1,21):
    data[f'Year {i}'] = None

last_quarter = 'CQ42019'

for i in range(1, 21):
    year_from_funds_life = f'Year {i}'
    
    # Calculating the average performance for the specified quarters
    data[year_from_funds_life] = data.apply(lambda row: np.nanmean(row.loc[f'CQ1{min(row["VINTAGE"] + i - 1, int(last_quarter[-4:]))}':f'CQ4{min(row["VINTAGE"] + i - 1, int(last_quarter[-4:]))}']), axis=1)
```

Logic behind the code below:

fund age = 5

i = 6

fund age <= i - 1

vintage = 2015

vintage >= 2020 - 6

year column = Year 6

```python
for i in range(2, 20):
    year_column = f'Year {i}'
    data.loc[(data['Fund Age'] <= (i - 1)) & (data['VINTAGE'] >= (2020 - i)), year_column] = None

# Quartile ranking based on the latest performance data
data['Quartile'] = pd.qcut(data['CQ42019'], q = 4, labels = [4, 3, 2, 1])
```

# Visualising Fund Performance per Quartiles across their Lifecycle

```python
fig, ax = plt.subplots(figsize=(10, 6))
fig.suptitle("Performance Across Quartiles a Fund's Life", fontsize=16, fontweight='bold')

for q in range(1, 5):
    quartile_data = data[data['Quartile'] == q] 
    performance_means = [quartile_data[f'Year {i}'].mean() for i in range(1, 11)]          # Plotting performance for each year
    ax.plot(range(1, 11), performance_means, label = f'Quartile {q}', marker = 'o', linewidth = 2)

ax.set_xlabel('Fund Lifecycle (Years)', fontsize = 14)
plt.xticks(np.arange(1, 11, 1))
ax.set_ylabel('Average MOIC', fontsize = 14)
ax.legend(fontsize = 12, title_fontsize = 12)
ax.grid(True, linestyle = '--', alpha = 0.7)
ax.set_axisbelow(True)

plt.tight_layout()
plt.show()
```
